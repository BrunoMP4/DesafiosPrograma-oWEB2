```python
import pandas as pd

uri = 'https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula1.2/ratings.csv'

df = pd.read_csv(uri)

filme_32 = df[(df['movieId'] == 32) & (df['rating'] > 3)]

if len(filme_32) > 0:
    print("Filme(s) com ID=32 e nota maior que 3:")
    for index, row in filme_32.iterrows():
        print("- ID:", row['movieId'])
        print("- Nota:", row['rating'])
        print()
else:
    print("Não há dados para o filme com ID=32 ou a nota é menor ou igual a 3.")
```
